```python
import serial
sourcePort="COM9"  #GPS rx port
destPort="COM17"   #sealink port
```

```python
gpsDataIN = serial.Serial(sourcePort,115200)

gpsDataOut = serial.Serial(destPort,115200)
```

```python
data = gpsDataIN.readline();
print(data)
```

```python
gpsDataOut.write(data)
```

```python
gpsDataIN.close()
gpsDataOut.close()
```

```python
# Script to convert HTML file to Arduino Serial.print(F("...")) statements and save to a file
with open('salaryslip2.html', 'r') as file:
    lines = file.readlines()

# Open the output file to write Arduino code
with open('output.html', 'w') as output_file:
    # Write a header comment for clarity
    output_file.write('// Arduino code generated from network_config.html\n\n')
    output_file.write('void sendWebPage() {\n')

    # Process each line of the HTML file
    for line in lines:
        # Escape double quotes and strip trailing whitespace
        escaped_line = line.strip().replace('"', '\\"')
        # Write the Serial.print statement to the file
        output_file.write(f'  g_confclient.print(F("{escaped_line}\\n"));\n')

    # Close the function
    output_file.write('}\n')
```
